# Practica 3: Neourona de McCulloch Pitts
### El modelo de McCulloch-Pitts (MCP) es una simplificación de una neurona, que toma varias entradas, las suma con un peso y luego aplica una función de activación (generalmente una función escalón) para determinar la salida.
### En esta practica entrenaremos una célula de McCulloch-Pitts para que sea capaz de comportarse como una compuerta lógica NOT de dos entradas

## Paso 1: Importar las librerias necesarias

```python
#Librerias
import numpy as np
```

## Paso 2: Definir la función de activación para el NOT lógico

```python
#Definir la función para el OR lógico

def McCulloch_Pitts_NOT(x1):
  #Pesos de las entradas
  w1 = -2

  #Umbral
  teta = -0.5

  #Suma ponderada de las entradas

  weighted_sum = x1*w1

  #salida de la célula

  if (weighted_sum >= teta):
    return 1
  else:
    return 0
```

## Paso 3: Probar la Neurona de McCulloch-Pitts para el OR lógico

```python
#Ejemplo de uso
print(McCulloch_Pitts_NOT(0))
print(McCulloch_Pitts_NOT(1))
```
